## sql

SQL (Structured Query Language) is a standard programming language used for managing and manipulating relational databases. It allows users to perform various operations on data stored in databases, such as querying, updating, deleting, and inserting data.

A relational database is a type of database that organizes data into tables (or relations) composed of rows and columns. It uses a structured schema to define how data is stored, accessed, and related to other data.

## Data Querying

Use SELECT statements to retrieve specific data from a database.

## DDL (Data Definition Language)
DDL stands for Data Definition Language, which is a subset of SQL used to define, modify, and delete database structures such as tables, schemas, indexes, and constraints. DDL statements do not deal with data itself but with the schema and structure of the database.

## Referential actions
Referential actions in a foreign key define how the database system should handle changes (like deletions or updates) to a referenced row in the parent table when that row has dependent rows in a child table. These actions help maintain referential integrity between the related tables.

## CASCADE
Automatically propagates the changes in the parent table to the child table.

ON DELETE CASCADE: Deletes all rows in the child table that reference the deleted row in the parent table.

ON UPDATE CASCADE: Updates the foreign key in the child table to match changes in the parent table's primary key.

## SET NULL

Sets the foreign key in the child table to NULL if the referenced row in the parent table is deleted or updated.

The foreign key column in the child table must allow NULL.

FOREIGN KEY (parent_id) REFERENCES parent(id) ON DELETE SET NULL ON UPDATE SET NULL;


## SET DEFAULT   ( not supported in MySQL)

Sets the foreign key in the child table to its default value if the referenced row in the parent table is deleted or updated.

The foreign key column in the child table must have a default value defined.

FOREIGN KEY (parent_id) REFERENCES parent(id) ON DELETE SET DEFAULT ON UPDATE SET DEFAULT;


## NO ACTION
Behavior: When you attempt to delete or update a parent row, the database checks if any child rows exist that reference it.

Enforcement Timing: The integrity check is deferred until the end of the current transaction.

This means: You might temporarily violate the foreign key constraint during the transaction.

If the violation still exists when the transaction ends (e.g., via COMMIT), the database raises an error and rolls back the transaction.


## RESTRICT
Behavior: The database immediately checks for referential integrity when you try to delete or update a parent row. If any dependent rows exist in the child table, the operation fails instantly.

Enforcement Timing: The integrity check is not deferred. 

This means: You cannot even temporarily violate the constraint within a transaction.
